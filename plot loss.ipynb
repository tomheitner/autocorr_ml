```python
%matplotlib widget
import matplotlib.pyplot as plt

import pickle

with open('./progress_log/pos_loss.pkl', 'rb') as f:
    pos_losses = pickle.load(f)
    
with open('./progress_log/neg_loss.pkl', 'rb') as f:
    neg_losses = pickle.load(f)

plt.figure(figsize=(10, 10))
ax1 = plt.subplot(2, 2, 1)
plt.plot(pos_losses)
plt.title('Loss for positive examples')
plt.grid()

plt.subplot(2, 2, 2, sharex=ax1, sharey=ax1)
plt.plot(neg_losses)
plt.title('Loss for negative examples')
plt.grid()

ax2 = plt.subplot(2, 2, 3, sharex=ax1)
plt.semilogy(pos_losses)
plt.title('Loss for positive examples')
plt.grid()

plt.subplot(2, 2, 4, sharex=ax1, sharey=ax2)
plt.semilogy(neg_losses)
plt.title('Loss for negative examples')
plt.grid()
```
